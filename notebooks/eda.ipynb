"Imported yolov8m and the LVIS dataset, the model will train during 300 epochs unless it sees no improvement in 100 consecutive iterations."

```python

model = YOLO("yolov8m.pt")

results = model.train(data= "../datasets/LVIS_Fruits_And_Vegetables/data.yaml", epochs = 300, imgsz = 640, device= 0)
```

"Continue the training if was cancelled"
